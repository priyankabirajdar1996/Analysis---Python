*Histogram* Visualize the Distribution of the Category Column

```python
df['Category'] = df['Category'].astype('str')
c_d =df['Category']
plt.figure(figsize=(10,6))
plt.hist(c_d,bins = 15,edgecolor='Black',color='Pink')
plt.xticks(rotation=90)
plt.show()
```

```python
# Most of the buyers are set
# Check b2b data
check_B2B=df['B2B'].value_counts()

# create pie chart
plt.pie(check_B2B,labels=check_B2B.index,autopct='%1.1f%%')
plt.show()
```

Maximum 99.3% of buyers are retailers and 0.8 % are Wholesalers

Available size as per category

```python
# Scatter plot
x_data = df['Category']
y_data = df['Size']
plt.scatter(x_data,y_data,color='Purple')
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Available size')
plt.show()
```

Statewise Total Order count

```python
top10_state = df['ship-state'].value_counts().head(10)
plt.figure(figsize=(30,10))
sns.countplot(data=df[df['ship-state'].isin(top10_state.index)],x="ship-state")
plt.xlabel('State') # Corrected from xlable to xlabel
plt.ylabel('Order_count') # Corrected from ylable to ylabel
plt.title('Distribution of state')
plt.xticks(rotation=90)
plt.show()
```

```python
# correlation matrix ( Heat Map)
correlation_matrix = df.select_dtypes(include=np.number).corr() # Select only numeric columns for correlation calculation
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Top 5 Cities Saleswise
city_sales = df.groupby('ship-city')['Amount'].sum().reset_index().sort_values(by=['Amount'], ascending=False).head(5)
top5_cities = df.groupby('ship-city')['Amount'].sum().sort_values(ascending=False).head(5)
print(top5_cities)
```
